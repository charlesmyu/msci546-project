```python
import pandas as pd
import csv
import os
```

```python
df = pd.read_csv("./raw/census_starter.csv")
df.head()
```

```python
# Transform into 4 files (YEAR.csv)
years = ['2017', '2018', '2019', '2020']
columns = ['pct_bb_', 'pct_college_', 'pct_it_workers_', 'median_hh_inc_']
updated_columns = ['pct_broadband', 'pct_college', 'pct_in_it_industry', 'median_hh_inc']

for year in years:
    rename_dict = {}
    for idx, col in enumerate(columns):
        rename_dict[col + year] = updated_columns[idx]

    original_columns = ['cfips'] + [column + year for column in columns]
    temp = df.filter(items=original_columns)
    temp = temp.rename(columns=rename_dict)
    temp.to_csv('./parsed/' + year + '.csv', index=False)
```
